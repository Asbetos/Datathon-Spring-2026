## Load All 5 Datasets

```python
from utils.data_loader import load_all, save_clean_parquet

data = load_all()
print(f"Loaded {len(data)} datasets: {list(data.keys())}")
```

## Quick Validation

For each dataset: shape, year range, null count. For state datasets: unique state count.

```python
for name, df in data.items():
    years = sorted(df['year'].unique())
    print(f"{name}: {df.shape}, years {years[0]}-{years[-1]}, nulls={df.isnull().sum().sum()}")
    if 'state' in name:
        print(f"  Unique states: {df['state'].nunique()}")
```
